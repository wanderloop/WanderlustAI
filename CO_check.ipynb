```
from google.colab.patches import cv2_imshow
from cv2 import imread, IMREAD_COLOR, rectangle, circle, waitKey, destroyAllWindows
img = imread('/content/1.jpg',
             IMREAD_COLOR,)

# Left=892, Top=458, Right=982, Bottom=706

def make_dots(x1, y1,
              x2, y2,):
    rectangle(img,
              (x1,y1),
              (x2,y2),
              (0,0,255),
               5,)
    a=(x1+x2)/2
    print(a)
    b=(y1+y2)/2
    print(b)
    circle(img,
           (int(a),
            int(b)),
            10,
            (0,255,0),
             -1,)
make_dots(175,
          540,
          238,
          703,)

cv2_imshow(img)
    
waitKey(0)
destroyAllWindows()
```
