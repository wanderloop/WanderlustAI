```
from google.colab.patches import cv2_imshow
from cv2 import imread, IMREAD_COLOR, rectangle, circle, imwrite, waitKey, destroyAllWindows
from PIL.Image import new

img = imread('/content/1.jpg',
             IMREAD_COLOR,)

print(img.shape)


# Left=892, Top=458, Right=982, Bottom=706


def make_dots(x1, y1,
              x2, y2,):
    rectangle(img,
             (x1,y1),
             (x2,y2),
             (0,0,255),
              5,)
    a=(x1+x2)/2
    print(a)
    b=(y1+y2)/2
    print(b)
    circle(img,
           (int(a),
            int(b)),
            10,
            (0,255,0),
             -1,)

x1=1156
y1=235
x2=1174
y2=280

# make_dots(175, 540,
#           238, 703,)

# make_dots(358, 536,
#           429, 682,)

make_dots(x1, y1,
          x2, y2,)
# cv2_imshow(img)

watch_face = img[y1:y2,
                 x1:x2,]
print(watch_face.shape)
img1 = new('RGB',
           (4096, 4096),
           color = 'white',)
img1.save('img1.jpg')
img1 = imread('img1.jpg')
img1= watch_face
print(img1.shape)
cv2_imshow(img)
cv2_imshow(img1)
imwrite('img1.jpg',
         img1,)

# Wait for user input before closing the image displayed
waitKey(0)
destroyAllWindows()
```
